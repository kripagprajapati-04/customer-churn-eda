```python
# step-1: IMPORTING LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# show plots inside the notebook

%matplotlib inline
```

```python
# Step-2 : LOAD THE DATASET
df = pd.read_csv('Customer Churn.csv')

# To show the first five columns
df.head()
```

```python
# What's the shape of data? (number of rows and column)
print('Numbers of Rows and Columns:', df.shape)

# BASIC INFO ABOUT EACH COLUMN
df.info()
```

```python
# SUMMARY STATISTICS FOR NUMERICAL COLUMNS
df.describe()
```

```python
# Summary for all columns (including object types)
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# convert into correct type
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
df['MonthlyCharges'] = pd.to_numeric(df['tenure'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['tenure'], errors='coerce')
```

```python
# Check how many rows have NaN
df['TotalCharges'].isnull().sum()
df['MonthlyCharges'].isnull().sum()
df['tenure'].isnull().sum()
```

```python
# Plot Churn Count
df ['Churn'].value_counts().plot(kind = "bar", color = ['salmon', 'skyblue'])
plt.title('Churn Distribution')
plt.xlabel("Churn")
plt.ylabel("Number of customers")
plt.xticks(rotation = 0)
plt.show()
                                 
```

```python
# Plotting Gender Distribution
df['gender'].value_counts().plot(kind = 'bar', color = ['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('GENDER') 
plt.ylabel("Number of people")
plt.xticks(rotation=0)
plt.show()                                
```

```python
# Monthly Charges Distribution(Numerical)
plt.hist(df['MonthlyCharges'], bins = 30, color = "teal")
plt.title("Distribution of monthly Charges")
plt.xlabel('Monthly Charges')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.hist(df['tenure'], bins=30, color='darkgreen')
plt.title('Customer Tenure Distribution')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Churn Rate by Contract Type
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack().plot(kind= 'bar', stacked = True, color = ['pink', 'magenta'])
plt.title('Churn Rate by Contract Type')
plt.ylabel('Proportion of Customers')
plt.xticks(rotation=0)
plt.show()
```

```python
# Do male and female customers churn at different rates?

# Customer Churn by Gender!
df.groupby('gender')['Churn'].value_counts(normalize=True).unstack().plot(
    kind='bar',
    stacked=True,
    color=['skyblue', 'salmon']
)
plt.title('Churn Rate by Gender')
plt.ylabel('Proportion of Customers')
plt.xticks(rotation=0)
plt.show()

# Churn rates were almost identical across male and female customers, indicating gender was not a significant factor in customer retention.”

```

Boxplot of Monthly charges vs churn

```python
df.boxplot(column = 'MonthlyCharges', by = 'Churn' , grid = True, patch_artist = True)
plt.title('Monthly charges vs churn')
plt.suptitle('')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```
